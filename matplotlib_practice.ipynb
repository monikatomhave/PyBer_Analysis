```python
#Import dependancies
import matplotlib.pyplot as plt
```

```python
#an array or list of data 5.1.3
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
```

```python
# using object-oriented interface method
# ax.plot() versus plt.plot()

#create the plot with ax.plt()
# plt.subplots() function returns a tuple that contains a figure and axes objects
# fig, ax = plt.subplots() unpacks the tuple into the variables fig and ax
# variable fig references the figure
# variable ax references the axes of the figure and annotates the figure
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
# seperate/replace fig, ax = plt.subplots() with fig = plt.figure() and ax = fig.add_subplot()

fig = plt.figure()
ax = fig.add_subplot()

# the result is an empty chart

# need to add the line following to get earlier result
ax.plot(x_axis, y_axis)
```

```python
# to create a quick graph using the object orientated interface method
# create the plot with as.plt()

ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# To change figure-level attributes (such as axis labels, a title, or a legend) or save the figure
# as an image, use fig = plt.figure()

# The plt.show() funtion looks for all of the active figure objects and opens a window that displays
# them if you have two or more sets of data to plot.

plt.plot(x_axis, y_axis)
plt.show()

# The plt.show() should only be used once per session because it can cause the graph to display
# improperly.  May need to restart the kernel and clear output. 
```

```python
#Annotate charts using Object-orientated Interface
#slightly different syntax
#  Matplotlib Object-Oriented Functions       Feature
# fig, ax = plt.subplots(figsize=(w, h))      Change the size of the figure in pixels. Add this in the subplots() function.
# ax.plot(x, y, label='line')                 Add a label that will be added to the legend.
# ax.set_ylim(min, max)                       Sets the min and max range of the y-axis.
# ax.set_xlim(min, max)                       Sets the min and max range of the x-axis.
# ax.set_xlabel('x label')                    Add a label to the x-axis.
# ax.set_ylabel('y label')                    Add a label to the y-axis.
# ax.set_title("Title")                       Add a title.
# ax.legend()                                 Add a legend.
# ax.grid()                                   Add a grid to the chart.
#** plt.savefig("add a path and figure extension")
#                                             Saves the figure with the given extension. Added at the end of your script.

# skill drill
import matplotlib.lines as mlines
fig, ax = plt.subplots()
# Create the plot.change width and color of the line and add markers for the data points.
ax.plot(x_axis, y_axis, marker="D", color="g", linewidth=2)
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a grid.
ax.grid()
# Add the legend.
label = mlines.Line2D([],[], color = 'green', marker = 'D', markersize = 5, label ='Boston')
ax.legend(handles = [label])
```

```python
#  %.1.7 create scatter plots and bubble charts
# used for regression analysis

# total number of rides
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
# Average Fare
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]
```

```python
# s must have same number of count as the arguments, or set to x to be a different size depending
# the value for that point, ie; s = 100 or s = x, etc

plt.scatter(x,y, c = 'green', marker = 'x', s = x)
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Average Fare vs Total Rides')
plt.show()
```

```python
# create a scatter plot using the MATLAB Method

# there are two ways, use either, plt.plot(), plt.scatter()

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
# adding the 'o' as a parameter inside the parenthesis will switch the plot from a line chart to a scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# 5.1.7 skill drill
# change color of marker to red, legend to Chicago, add title and axes labels
# switch axes so Fare($) data is on x-axis
# add limit for the x-axis data
# invert the y-axis so January is on top
import matplotlib.pyplot as plt
plt.scatter(y_axis, x_axis, c = 'red', label = 'Chicago')
plt.xlim(0, 50)
plt.xlabel('Fare ($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.legend()
plt.gca().invert_yaxis()
plt.show()
```

```python
#create a bubble chart using the MATLAB method
# s is the size parameter for the marker and must be equal to a feature of the data
# ie, we make the size of each marker equal to the data floating-point decimal values of the y-axis

plt.scatter(x_axis, y_axis, s = y_axis)
plt.show()
```

```python
# we  can adjust the size by multiplying the data in the y-axis value.
# have to iterate through the y-axis data and multiply each data point and add it to the list

y_axis_larger = []

for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s = y_axis_larger)

plt.show()
```

```python
# we can refactor the code but use Python's list comprehension technique inside the plt.scatter() function.
# use list comprehension to replace for or while loops, which is faster because it is optimized for 
# python to spot predictable patterns during looping.
# new_list = [expression for item in list if conditional]

plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
plt.show()
```

```python
# create a scatter plot using Object-Oriented Interface

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# create a bubble chart using OO interface
# add s parameter just like we did using the MATLAB

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s = y_axis)
```

```python
# skill drill using OO appraoch, add changes to bubble chart
# change color of markers to blue
# change size of markers to 5 times data point
# make color 20% transparent
# add a black egde color to the circles
# make linewidth of circles 2 points
# add a legend for the city of Boston
# add a title
# switch axis for fares is on x-axis
# add a limit for the x-axis
# invert y-axis so january is at the top

fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i*5 for i in y_axis], c = 'blue', label = 'Boston', alpha = .2, linewidths = 2, edgecolors = 'k')
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.set_xbound(0,50)
ax.legend()
ax.invert_yaxis()
```

```python
# 5.1.8 pie charts
# video clip example

categories = ['pants', 'shirts', 'outerwear', 'pajamas', 'underwear', 'shoes']
revenue = [25, 15, 5, 20, 10, 25]
#create pie plot, only takes one madatory argument, size of each slice
plt.pie(revenue)
plt.show()
```

```python
# add labels to each slice

plt.pie(revenue, labels = categories)
plt.show()
```

```python
# put exact percentages in each category

plt.pie(revenue, labels = categories, autopct = "%1.1f%%")
plt.show()
```

```python
# specific category to highlight, using explode from the chart
# shirt is second in the cateogries list, so create an explode list to match number of values
explode = [0, .1, 0, 0, 0, 0]
plt.pie(revenue, labels = categories, autopct = "%1.1f%%", explode = explode)
plt.show()
```

```python
# make pic chart more visually interesting

# Assign colors to each wedge, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values = (0, 0, 0, 0, 0, 0, .2, 0, 0, 0, 0, 0)

# percentages are crowded so increase size of pie chart, added after running other three lines
plt.subplots(figsize = (8, 8))

plt.pie(y_axis, explode = explode_values, colors = colors, labels = x_axis, autopct = '%.1f%%')
plt.show()
```

```python
# 5.1.8 Skill drill
# add a percentage to one decimal place to each wedge
# increase figure to 8X8
# explode two highest percentage months
# add a shadow
# add a start angle so January is at the top
# Reverse the order so months go in clockwise direction
# add new colors for each wedge

fig, ax = plt.subplots(figsize = (8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values = (0, 0, .2, 0, 0, 0, .2, 0, 0, 0, 0, 0)

ax.pie(y_axis, explode = explode_values, labels = x_axis, colors = colors, autopct = '%.1f%%', shadow = True, startangle = 97, counterclock = False)
plt.show()
```
